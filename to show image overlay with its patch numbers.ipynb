```python
import cv2

image_path = "input_image.jpg"
patch_size = 16

img = cv2.imread(image_path)
overlay = img.copy()

h, w, _ = img.shape
patch_id = 0

for y in range(0, h - patch_size + 1, patch_size):
    for x in range(0, w - patch_size + 1, patch_size):
        # Draw rectangle
        cv2.rectangle(
            overlay,
            (x, y),
            (x + patch_size, y + patch_size),
            (0, 255, 0),
            1
        )

        # Put patch index (small font)
        cv2.putText(
            overlay,
            str(patch_id),
            (x + 1, y + 12),
            cv2.FONT_HERSHEY_PLAIN,
            0.6,
            (0, 0, 255),
            1
        )

        patch_id += 1

cv2.imshow("Patch Index Overlay", overlay)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
